```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import plotly as py
from datetime import datetime
from matplotlib.pyplot import figure
import plotly.express as px
import openpyxl
from openpyxl import Workbook
import os
from matplotlib.pyplot import figure
import scipy.stats as st
os.getcwd()
```

```python
#Import data 

data = pd.read_excel('data_daily.xlsx', engine ='openpyxl')
```

```python
#Set time as index 

data = data.set_index('dates')
```

```python
#Verify data import

figure(figsize=(32, 16), dpi=80)
plt.plot(data['spx'], label='spx')
plt.plot(data['dax'], label='dax')
plt.plot(data['hsi'], label='hsi')
plt.plot(data['hsci'], label='hsci')
plt.plot(data['btc'], label='btc')
plt.plot(data['oil'], label='oil')
plt.plot(data['yield2y'], label='yield2y')
plt.plot(data['yield10y'], label='yield10y')
plt.plot(data['xau'], label='xau')
plt.legend(loc='best')
plt.show()
```

```python
#Calculate rolling window correlation coeficient of one week

week_window_SPX_DAX = data['dax'].rolling(5).corr(data['spx'])
week_window_SPX_HSI = data['hsi'].rolling(5).corr(data['spx'])
week_window_SPX_HCSI = data['hsci'].rolling(5).corr(data['spx'])
week_window_SPX_BCT = data['btc'].rolling(5).corr(data['spx'])
week_window_SPX_OIL = data['oil'].rolling(5).corr(data['spx'])
week_window_SPX_YIELD2Y = data['yield2y'].rolling(5).corr(data['spx'])
week_window_SPX_YIELD10Y = data['yield10y'].rolling(5).corr(data['spx'])
week_window_SPX_XAU = data['xau'].rolling(5).corr(data['spx'])
```

```python
#Plot correlations

week_windows = [
(week_window_SPX_DAX, 'dax'),
(week_window_SPX_HSI, 'hsi'),
(week_window_SPX_HCSI, 'hcsi'),
(week_window_SPX_BCT, 'btc'),
(week_window_SPX_OIL, 'oil'),
(week_window_SPX_YIELD2Y, 'yield2y'),
(week_window_SPX_YIELD10Y, 'yiedl10y'),
(week_window_SPX_XAU, 'xau')      
]


for week_window in week_windows:
    dataframe, ticker = week_window
    fig = px.line(dataframe, title=f"Correlation Coeficient Week Window spx - {ticker}")
    fig.show()
```

```python
#Plot Correlations together sample of June 2022

figure(figsize=(32, 16), dpi=80)
plt.plot(week_window_SPX_DAX['2022-06'], label='dax')
plt.plot(week_window_SPX_HSI['2022-06'], label='hsi')
plt.plot(week_window_SPX_HCSI['2022-06'], label='hsci')
plt.plot(week_window_SPX_BCT['2022-06'], label='btc')
plt.plot(week_window_SPX_OIL['2022-06'], label='oil')
plt.plot(week_window_SPX_YIELD2Y['2022-06'], label='yield2y')
plt.plot(week_window_SPX_YIELD10Y['2022-06'], label='yield10y')
plt.plot(week_window_SPX_XAU['2022-06'], label='xau')
plt.legend(loc='best')
plt.show()
```

```python
#Notes:

#After analyzing peaks con price percentage correlation, for extreme peaks, we have a 5 day rolling 
#window with a really low correlation, then correlation increases significantly and then the percentage change 
#is extreme 

#Usually extreme changes are not noisy and do not offer any significant insight 
```

```python
#Plot distribution of correlations

for week_window in week_windows:
    dataframe, ticker = week_window
    fig = px.histogram(dataframe, title=f"Correlation Coeficient Week Window spx - {ticker}")
    fig.show()
```

```python
#Calculate 95% confidence intervals for weekly correlation rolling window  

for week_window in week_windows:
    dataframe, ticker = week_window
    print(f"95% confidence interval of correlation between from spx - {ticker} on a weekly rolling window")
    interval = st.t.interval(alpha=0.95, df=len(dataframe.dropna()), loc=np.mean(dataframe.dropna()), scale=st.sem(dataframe.dropna())) 
    print(interval)
    
```

```python
#Example of exponential moving avarage construction

figure(figsize=(16, 8), dpi=80)

#build ema
ema = week_window_SPX_DAX.ewm(com=4).mean()

#add ema 

plt.plot(ema['2022'], color = 'r', linestyle = ':', label = "ema of 4 periods on 1 week rolling window correlation")

#actual plot

plt.plot(week_window_SPX_DAX['2022'], label='spx-dax')
plt.legend(loc='best')
plt.show()
```
